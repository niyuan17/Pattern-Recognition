```python
import pandas as pd
import numpy as np
```

```python
# read index constituents
idx_ctt = pd.read_csv(r".\Data\sp500_constituents_since1980.csv")
sp500_ticker = idx_ctt['co_tic']
sp500_ticker.to_csv(".\data\sp500_tickers_since1980.csv", index = False)
sp500_ticker = sp500_ticker.tolist()
sp500_ticker = [str(x).split('.')[0] for x in sp500_ticker]
sp500_ticker = list(set(sp500_ticker))
print len(sp500_ticker)
```

```python
# read index constituents
idx_ctt = pd.read_csv(r".\Data\sp500_constituents_since1980.csv")
sp500_cusip = idx_ctt['co_cusip']
sp500_cusip.to_csv(".\data\sp500_cusip_since1980.csv", index = False)
sp500_cusip = sp500_cusip.tolist()
sp500_cusip = list(set(sp500_cusip))
print len(sp500_cusip)
```

```python
# Seperate stocks by ticker
chunksize = 10**5
iii = 0
for chunk in pd.read_csv(r".\Data\stock price\sp500stocks_19802016.csv", chunksize = chunksize, index_col = 1, parse_dates = True):
    del chunk['PERMNO']
    columns_= chunk.columns.values
    for this_ticker in sp500_ticker:
        this_com = chunk[chunk['TICKER'] == this_ticker]
        this_com = this_com[columns_]
        if len(this_com) > 0:
            print 'find', this_ticker
            filepath = r'.\Data\stock price\sp500_80_16\%s.csv'% (this_ticker)
            try:
                tmp = pd.read_csv(filepath, index_col = 0, parse_dates = True)
                tmp = tmp[columns_]
                tmp = pd.concat([tmp,this_com])
                tmp.to_csv(filepath)
            except:
     
                this_com.to_csv(filepath)
```
